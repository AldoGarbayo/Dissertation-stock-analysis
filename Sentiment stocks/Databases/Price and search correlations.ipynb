```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

```

```python
#American Airlines Correlations
AAL_data = pd.read_csv('AAL Weekly Trend .csv', index_col = 'Week')
AAL_data.head()
```

```python
AAL_data.index = pd.to_datetime(AAL_data.index)
AAL_data.head()
```

```python
AAL_data['Trend'].plot(label = 'Trend',figsize = (16,8), title = 'Trend and stock movements')
AAL_data['Price'].plot(label = 'Price')
plt.legend()
```

```python
Price = AAL_data['Price']
Trend = AAL_data['Trend']
```

```python
AAL_data.corr()
```
